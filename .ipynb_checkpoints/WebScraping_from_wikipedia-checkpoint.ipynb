```python
import requests
from bs4 import BeautifulSoup
import html5lib
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Asian_countries_by_area"
response = requests.get(url)
str_data = response.text

#now parse the data return from the website
tags = BeautifulSoup(str_data, 'html.parser')
tags.title
```

```python
tags.title.string
```

```python
tags.a
```

```python
a_links = tags.find_all('a')
for link in a_links:
    print(link.get('href'))
```

```python
all_tables = tags.find_all('table')
all_tables
```

```python
right_table = tags.find('table', class_='wikitable sortable jquery-tablesorter')
right_table
```
