#### histogram:
- it is used on one variable
- on x axis: range and y-axis: frequency

```python
iris.head()
```

```python
print(max(iris['sepal_length']))
print(min(iris['sepal_length']))

print("range:",7.9-4.3)            # understand the range is imp
print("bucket size:",(7.9-4.3)/10)       # bucket size
print("min value + bucket size gives next value:",4.3+0.36)
```

```python
plt.hist(iris['sepal_length'])           #bydefault bins=10      #    iris['sepal_length'].plot.hist()
plt.xlabel('range')
plt.ylabel('Frequency')
```

```python
iris.plot.hist()                   #  iris.plot(kind = 'hist')
```

- In above graph we not able to see the background data
- stacked means bar one on the another 
- by using the bins we can control width of the graph

```python
iris.plot(kind = 'hist', stacked = False, bins = 100)
```

- To present the graph orientation='horizontal'

```python
iris.plot(kind = 'hist', stacked = True, bins = 50, orientation = 'horizontal')  
```

- in between what is max value of data?
- from 0.0 to 0.5 we have max value of data i.e 15.8
- 2nd max data is in b/w -0.5 to 0.0 i.e 15

```python
iris['sepal_width'].diff()
```

```python
iris['sepal_width'].diff().plot(kind = 'hist', stacked = True, bins = 50)
```
